```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("steam.csv")

# Basic inspection
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns.tolist())

print("\nFirst 5 rows:")
print(df.head())

print("\nMissing values:")
print(df.isnull().sum().sort_values(ascending=False).head(20))
```

```python
selected_cols = [
    "name",
    "release_date",
    "english",
    "developer",
    "publisher",
    "platforms",
    "required_age",
    "categories",
    "genres",
    "steamspy_tags",
    "achievements",
    "positive_ratings",
    "negative_ratings",
    "average_playtime",
    "median_playtime",
    "owners",
    "price"
]

df = df[selected_cols].copy()
df = df.drop_duplicates()

print("New shape after selecting columns and dropping duplicates:", df.shape)
df.head()
```

```python
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")

df["total_reviews"] = df["positive_ratings"] + df["negative_ratings"]
df["positive_ratio"] = df["positive_ratings"] / df["total_reviews"].replace(0, np.nan)
df["release_year"] = df["release_date"].dt.year
df["is_free"] = (df["price"] == 0).astype(int)

current_year = 2026
df["game_age"] = current_year - df["release_year"]

print(df[["total_reviews", "positive_ratio", "release_year", "is_free", "game_age"]].head())
```

```python
df = df.dropna(subset=["price", "positive_ratings", "negative_ratings", "release_year"])

print("Final shape:", df.shape)
print(df[["price", "positive_ratings", "negative_ratings", "total_reviews", "positive_ratio", "release_year", "game_age"]].describe())
```

```python
df.to_csv("steam_cleaned.csv", index=False)
print("steam_cleaned.csv created successfully.")
```
