```python
import pandas as pd
from scipy.stats import ttest_ind, pearsonr

df = pd.read_csv("steam_cleaned.csv")

print("Shape:", df.shape)
print(df[["price", "positive_ratio", "is_free", "genres", "total_reviews"]].head())
```

# Hypothesis Testing

This notebook tests whether selected game-related variables are significantly associated with video game success, measured mainly by positive review ratio.

## Hypothesis 1: Free vs Paid Games

**H0:** Free-to-play and paid games have the same mean positive review ratio.  
**H1:** Free-to-play and paid games have different mean positive review ratios.

I will use an independent samples t-test.

```python
free_games = df[df["is_free"] == 1]["positive_ratio"].dropna()
paid_games = df[df["is_free"] == 0]["positive_ratio"].dropna()

t_stat, p_value = ttest_ind(free_games, paid_games, equal_var=False)

print("Hypothesis 1: Free vs Paid Games")
print("Free games count:", len(free_games))
print("Paid games count:", len(paid_games))
print("Mean positive ratio (free):", free_games.mean())
print("Mean positive ratio (paid):", paid_games.mean())
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

Since the p-value is greater than 0.05, the null hypothesis cannot be rejected. This means that there is no statistically significant difference between free-to-play and paid games in terms of mean positive review ratio.

## Hypothesis 2: Price and Positive Review Ratio

**H0:** There is no linear relationship between price and positive review ratio.  
**H1:** There is a linear relationship between price and positive review ratio.

I will use Pearson correlation.

```python
temp = df[["price", "positive_ratio"]].dropna()

corr, corr_p = pearsonr(temp["price"], temp["positive_ratio"])

print("Hypothesis 2: Price and Positive Ratio")
print("Correlation coefficient:", corr)
print("p-value:", corr_p)
```

Since the p-value is greater than 0.05, the null hypothesis cannot be rejected. This suggests that no statistically significant linear relationship was found between price and positive review ratio.

## Hypothesis 3: Total Reviews and Positive Review Ratio

**H0:** There is no linear relationship between total review count and positive review ratio.  
**H1:** There is a linear relationship between total review count and positive review ratio.

I will use Pearson correlation.

```python
temp_reviews = df[["total_reviews", "positive_ratio"]].dropna()

corr_reviews, p_reviews = pearsonr(temp_reviews["total_reviews"], temp_reviews["positive_ratio"])

print("Hypothesis 3: Total Reviews and Positive Ratio")
print("Correlation coefficient:", corr_reviews)
print("p-value:", p_reviews)
```

Since the p-value is greater than 0.05, the null hypothesis cannot be rejected. There is no statistically significant linear relationship between total review count and positive review ratio.
